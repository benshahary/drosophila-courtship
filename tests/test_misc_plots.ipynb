```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as mpl_cm
%matplotlib inline

from context import (
    experiment,
    plots,
    misc_plots,
    plot_utils
)
```

```python
exp = experiment.FixedCourtshipTrackingExperiment.from_dir(
    data_dirs=['/Volumes/Ross_Seagate/Dropbox/ben-shahar_lab/midline crossing/behavior/control_courtship/fcts2']
)
```

```python
# plot the centroid-to-centroid distance during courtship (aka. the courtship path)
# the gray, dashed line represents the expected distance the male must be from the 
# female to engage in tapping.
misc_plots.plot_dists_cartesian(exp, colors=['k'])
```

```python
# plot the distance the male is from the expected
# tapping distance (gray line)
misc_plots.plot_residual_dists(exp, colors=['k'])
```

```python
# plot the average courtship path on polar axes
misc_plots.plot_dists_polar(exp, colors=['k'])
```

```python
# plot the average courtship path on polar axes and show the 
# outline of the female ellipse as a dashed line.
misc_plots.plot_dists_polar(exp, colors=['k'], draw_female_ellipse=True)
```

```python
# plot the average courtship path on polar axes, show the
# outline of the female ellipse, and draw the expected 
# tapping distance.
fig, ax = plt.subplots(subplot_kw=dict(polar=True))
misc_plots.plot_dists_polar(exp, colors=['k'], draw_female_ellipse=True, ax=ax)
misc_plots.plot_expected_tapping_distance(exp, ax)
```

```python
# plot the ratio of the maximal distance when male is on
# anterior half of female to when male is on posterior half.
misc_plots.plot_dists_peak_ratios(exp)
```

```python
# Note :: in this and all subsequent plots, behaviors are not 
#         mutually exclusive of one another because a behavioral
#         hierarchy was not created.

# plot the average tapping location
misc_plots.plot_arrow_polar_01(exp, 'tapping', 'control', spread=0.5)
```

```python
# plot the average orienting location
misc_plots.plot_arrow_polar_01(exp, 'orienting', 'control', spread=0.5)
```

```python
# plot the average scissoring location
misc_plots.plot_arrow_polar_01(exp, 'scissoring', 'control', spread=0.5)
```

```python
# plot the courtship index
misc_plots.plot_ci(exp, ['C0'])
```

```python
# plot the courtship latency
misc_plots.plot_cl(exp, ['C0'])
```

```python
# plot behavioral matrices
misc_plots.plot_behavioral_mats(
    exp, 
    'control', 
    ['courtship_gt', 'tapping', 'orienting', 'scissoring'], 
    cmap=mpl_cm.binary
)
```
